```python
# ============================================================================
# CELL 2: Download ODIR-5K from Kaggle
# ============================================================================
# Upload your kaggle.json first in Colab:
#   from google.colab import files; files.upload()
#   !mkdir -p ~/.kaggle && mv kaggle.json ~/.kaggle/ && chmod 600 ~/.kaggle/kaggle.json

%pip install -q --upgrade kaggle

import os
import subprocess
from pathlib import Path

DATA_ROOT = Path("/content/odir5k")
DATA_ROOT.mkdir(parents=True, exist_ok=True)

kaggle_json = Path.home() / ".kaggle" / "kaggle.json"
if not kaggle_json.exists():
    raise FileNotFoundError(
        f"Kaggle credentials not found at {kaggle_json}. Upload/place kaggle.json first."
    )

cmd = [
    "kaggle",
    "datasets",
    "download",
    "andrewmvd/ocular-disease-recognition-odir5k",
    "-p",
    str(DATA_ROOT),
    "--unzip",
]
subprocess.run(cmd, check=True)

# Find the actual data paths
for root, dirs, files in os.walk(DATA_ROOT):
    for f in files:
        if f.endswith((".csv", ".xlsx")):
            print(f"📄 Annotation file: {os.path.join(root, f)}")
    if len(dirs) == 0 and len(files) > 10:
        exts = set(f.split(".")[-1].lower() for f in files)
        if exts & {"jpg", "jpeg", "png"}:
            print(f"🖼️  Image directory: {root} ({len(files)} files)")
```

```python
# ============================================================================
# CELL 4: Download RETFound pretrained weights (from Hugging Face)
# ============================================================================
%pip install -q huggingface_hub

import os
import shutil
from pathlib import Path
from huggingface_hub import hf_hub_download

RETFOUND_PATH = Path("/content/RETFound_cfp_weights.pth")
RETFOUND_PATH.parent.mkdir(parents=True, exist_ok=True)

# Set FORCE_REDOWNLOAD = True if you want to re-download even if file exists
FORCE_REDOWNLOAD = False

if not RETFOUND_PATH.exists() or FORCE_REDOWNLOAD:
    # Remove old file if forcing re-download
    if RETFOUND_PATH.exists():
        RETFOUND_PATH.unlink()
        print("🗑️  Removed old weights file.")

    print("⏳ Downloading RETFound weights from Hugging Face...")
    cached_file = hf_hub_download(
        repo_id="rmaphoh/RETFound_MAE",
        filename="RETFound_cfp_weights.pth",
    )
    shutil.copy(cached_file, RETFOUND_PATH)
    size_mb = RETFOUND_PATH.stat().st_size / 1e6
    print(f"✅ RETFound weights downloaded: {size_mb:.0f} MB")
else:
    size_mb = RETFOUND_PATH.stat().st_size / 1e6
    print(f"✅ RETFound weights already present: {size_mb:.0f} MB")
    print(f"   Path: {RETFOUND_PATH}")
    print(f"   (Set FORCE_REDOWNLOAD = True to re-download)")
```
